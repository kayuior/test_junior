```python
exploded_posts_df = (
    posts_df.dropna(subset="tags")
    .explode("tags_in_list")
    .rename(columns={"tags_in_list": "tag"})
)
```

```python
exploded_posts_df = exploded_posts_df.merge(users_df,how="left",left_on="owner_user_id",right_on="id")
```

```python
exploded_posts_df["count"] = 1
```

```python
exploded_posts_df
```

```python
tag_list = exploded_posts_df["tag"].unique().tolist()
```

```python
len(set(tag_list))
```

```python
tags_df = exploded_posts_df.groupby(by = "tag").agg({"count":"sum","reputation":"mean","score":"mean"}).reset_index()
```

```python
tags_df = tags_df.rename(
    columns={
        "count": "posts",
        "reputation": "avg_reputation",
        "score": "avg_score",
    }
)
```
